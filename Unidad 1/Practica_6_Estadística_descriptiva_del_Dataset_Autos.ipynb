```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset de autos
dataset_url = "/content/autos2 (1).csv"
columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors",
           "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", "height",
           "curb-weight", "engine-type  ", "num-of-cylinders", "engine-size", "fuel-system", "bore",
           "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
df = pd.read_csv(dataset_url, names=columns, na_values="?")
```

```python
# Convertir la columna 'price' a tipo numérico
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
# Estadísticas descriptivas básicas
min_price = df['price'].min()
max_price = df['price'].max()
mode_price = df['price'].mode()[0]
mean_price = df['price'].mean()
median_price = df['price'].median()
std_price = df['price'].std()
```

```python
describe_stats = df['price'].describe()
```

```python
print(f"Valor mínimo de precios: {min_price}")
print(f"Valor máximo de precios: {max_price}")
print(f"Valor de la moda de precios: {mode_price}")
print(f"Valor de la media de precios: {mean_price}")
print(f"Valor de la mediana de precios: {median_price}")
print(f"Valor de la desviación estándar de precios: {std_price}")
print("\nEstadísticas generales:")
print(describe_stats)
```

```python
# Histograma de precios con la regla de Sturges
num_bins = int(1 + np.log2(len(df['price'].dropna())))
plt.figure(figsize=(10, 5))
plt.hist(df['price'].dropna(), bins=num_bins, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribución de Precios de los Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Conclusion del Punto No. 2:
##### El histograma muestra que la mayoría de los automóviles tienen precios en un rango accesible, con una distribución sesgada hacia los valores más bajos. Esto sugiere que existen más vehículos de gama media-baja en el mercado.

```python
# Boxplot Precio vs Body Style
plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=df)
plt.title("Relación entre Precio y Tipo de Vehículo")
plt.xlabel("Tipo de Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()
```

# Conclusión del Punto No.3:
##### Se observa que los diferentes tipos de carrocería presentan variaciones significativas en los precios. Los sedanes tienden a ser más económicos en comparación con otros estilos de carrocería como convertibles y hatchbacks, que tienen una mayor variabilidad de precios.

```python
# Boxplot Precio vs Drive Wheels
plt.figure(figsize=(10, 5))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title("Relación entre Precio y Tipo de Tracción")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
plt.show()
```

# Conclusión del Punto No.4:
##### Los vehículos con tracción en las cuatro ruedas suelen tener precios más altos en comparación con los de tracción delantera o trasera. Esto puede deberse a los costos adicionales en tecnología y rendimiento asociados a este tipo de tracción.

```python
# Boxplot Precio vs Engine Location
plt.figure(figsize=(10, 5))
sns.boxplot(x='engine-location', y='price', data=df)
plt.title("Relación entre Precio y Ubicación del Motor")
plt.xlabel("Ubicación del Motor")
plt.ylabel("Precio")
plt.show()
```

# Conclusión del Punto No.5:

##### Se puede notar que los vehículos con motor ubicado en la parte trasera suelen tener precios significativamente más altos. Esto sugiere que esta configuración está asociada con automóviles de alto rendimiento o deportivos.
