```python
# basado en el LIBRO para Administracion
# creamos una lista vacia que representa la edad de las personas
import random
edades = []
# hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range (100):
  # generamos un numero aleatorio con la funcion randint y lo guardamos en la variable
  numero_aleatorio = random.randint(18,78)
  # Agregamos el numero aleatorio a la lista de Valores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# Crear un Dataframe de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edades'])
df.head()
```

```python
# Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edades'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

# Imprimir la tabla de Distribucion de frecuencias
print("Tabla de Distribucion de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de Sturges

c = 1 + log2(N)

```python
# Importamos la libreria math de funciones matematicas
import math
```

```python
print (edades)
```

```python
# Crear un Dataframe con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el numero de intervalos utilizando la regla de Sturges
n = len(edades)
K = int(1 + math.log2(n))
print("Valor de n = ", n , "\nNumero de Intervalos =", K)
```

```python
# Crear la tabla de Frecuencias

tabla_frecuencias = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
# Ordenar la tabla de Frecuencia por el intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
# Generar las columnas siguientes
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencias)
```

# Practica 5: Diagramas de Cajas y bigotes

```python
import random
edades2 = []
for i in range (100):
  numero_aleatorio = random.randint(18,70)
  edades2.append(numero_aleatorio)
print(edades2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(edades,
                     vert=True, # Vertical box aligment
                     patch_artist=True) # fill with color
ax1.set_title('Edades Encuesta 1')

bplot2 = ax2.boxplot(edades2,
                     vert=True, # Vertical box aligment
                     patch_artist=True) # fill with color
ax2.set_title('Edades Encuesta 2')
plt.show()
```
