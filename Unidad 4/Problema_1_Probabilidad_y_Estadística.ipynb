#*Problema 1*
######Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:


---



###### -La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
###### -Cuantos estudiantes pesan más de 68 y menos de 72 kg.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500


datos = np .random.normal (media , desv_std , num_datos)
datos[0:500]
```

```python
# # Crear un rango de valores para crear una densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# # # Calcular la PDF de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media) ** 2 / (2 * desv_std ** 2))
```

```python
# Graficar Histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') # Histograma de los datos
plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los Estudiantes')
plt.grid(True)
plt.show()
```

```python
# Vizualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Peso en kg. de los Estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar Histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal')
plt.xlabel('Peso en kg de los Estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Parametros de la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion Estandar
total_estudiantes = 500

# VARIABLE para calculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 72

prob_menos_72kg = norm.cdf(peso_limite, media, desv_std)
```

```python
estudiantes_menos_72kg = total_estudiantes * prob_menos_72kg

print("Probabilidad de que un estudiante pese menos de 72 kg:",prob_menos_72kg)
print("Numero de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 72 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

```python
prob_menos_68kg = norm.cdf(peso_limite, media, desv_std)

prob_mas_68kg = 1 - prob_menos_72kg

estudiantes_mas_68kg = total_estudiantes * prob_mas_68kg

print("Probabilidad de que un estudiante pese mas de 68 kg:", prob_mas_68kg)
print("Numero aproximado de estudiantes que pesan mas 68 kg:", estudiantes_mas_68kg)
```

```python
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 68 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```

# Combinando las dos graficas nos da:

```python
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= 68), color='skyblue', alpha=0.5, label='Peso > 68 kg')
plt.fill_between(x, y, where=(x <= 72), color='skyblue', alpha=0.5, label='Peso < 72 kg')

estudiantes_mas_68kg = (total_estudiantes * prob_mas_68kg)
estudiantes_menos_72kg = (total_estudiantes * prob_menos_72kg)

print('-Probabilidad de que un estudiante pese mas de 68 kg:', prob_mas_68kg)
print('-Numero aproximado de estudiantes que pesan mas 68 kg:', estudiantes_mas_68kg)

print("-Probabilidad de que un estudiante pese menos de 72 kg:",prob_menos_72kg)
print("-Numero de estudiantes que pesan menos de 72 kg:", estudiantes_menos_72kg)

plt.title('Distribucion de pesos de estudiantes\n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
