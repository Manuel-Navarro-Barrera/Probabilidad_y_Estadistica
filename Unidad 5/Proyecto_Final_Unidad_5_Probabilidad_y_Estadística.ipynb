# Grupo 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy import stats
```

```python
grupo1 = pd.read_csv('grupo1.csv')
grupo1.head(50)
```

```python
print("Descripción estadistica Grupo 1")
print(grupo1.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(grupo1['Calificacion'], bins=15, kde=True)
plt.title('Distribución de Calificaciones-Grupo 1')
plt.xlabel('Calificacion')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.regplot(x='HorasEstudio', y='Calificacion', data=grupo1, ci=None)
plt.title('Relación entre Horas de Estudio y Calificación (Grupo 1)')
plt.ylabel('Horas de Estudio')
plt.xlabel('Calificaciones')
plt.show()
```

```python
x1= grupo1[['HorasEstudio']]
y1= grupo1[['Calificacion']]
model1 = LinearRegression()
model1.fit(x1, y1)
```

```python
# Coeficientes
print("\nModelo de Regresión-Grupo 1")
print(f"Intercepto (b0): {model1.intercept_[0]:.2f}")
print(f"Coeficiente (b1): {model1.coef_[0][0]:.2f}")
```

```python
corr1, p_value1 = stats.pearsonr(grupo1['HorasEstudio'], grupo1['Calificacion'])
print(f"\nCoeficiente de correlacion de Pearson: {corr1:.3f}")
print(f"Valor p: {p_value1:.4f}")
```

# Repuesta a las Preguntas

Captura de pantalla 2025-05-15 172019.png

```python
print("\nInterpretacion de Resultados - Grupo 1:")
print("1. El coeficiente de regresion (b1) indica que por cada hora adicional de estudio, la calificacion aumenta en promedio.",f"{model1.coef_[0][0]:.2f} puntos.")
print("2. El coeficiente de correlacion de Pearson (", f"{corr1:.3f}", ") sugiere una relacion" "postiva"
      if corr1 > 0 else "negativa", "y", "fuerte" if abs(corr1) > 0.7 else "moderada" if abs(corr1) > 0.3 else "debil", ") entre las variables.")
print("3. El modelo", "es" if p_value1 < 0.05 else "no es", "estadisticamente significativo.")
print("4. La confiabilidad del modelo para predecir calificaciones es", "alta" if abs(corr1) > 0.7 else "moderada" if abs(corr1) > 0.5 else "baja",
      "basado en el coeficiente de correlacion.")
```
